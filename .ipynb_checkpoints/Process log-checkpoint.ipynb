# Data Preperation

preparing (or pre-processing) raw data or disparate data sources into refined information.

```python
with open('log-file.log', 'rt') as f,open('outfile.txt', 'w') as o: # Open file log-file.log for reading of log data.
    for line in f:
        if  ("CompanyX OCR Region" in line) or ("Updating Siemens train info"  in line):
            print (line)
            o.write(line)
```

```python
import pandas as pd
import re
import datetime
import numpy as np

df = pd.DataFrame({'Iteration': [], 'Time': [], 'Success_flag': []})
new_detection=0
success_detection=0
missed_detection=0
flag=0

with open('outfile.txt', 'rt') as f:
    for line in f:
        if ('[TRACE]' in line) and (flag==0): #New Detection
            flag=1
            match = re.search('\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}', line)
            if match is not None:
                start_date = datetime.datetime.strptime(match.group(), '%Y-%m-%d %H:%M:%S')
            else:
                start_date=None
        elif ('[TRACE]' in line) and (flag==1): #Missed Detection
            flag=1
            missed_detection+=1
            new_detection+=1
            match = re.search('\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}', line)
            df = df.append({'Iteration': new_detection, 'Time': start_date, 'Success_flag': 0}, ignore_index=True)
            if match is not None:
                start_date = datetime.datetime.strptime(match.group(), '%Y-%m-%d %H:%M:%S')
            else:
                start_date=None
        elif ('[INFO ]' in line) and (flag==1): #Sucessfull Detection
            flag=0
            success_detection+=1
            new_detection+=1
            df = df.append({'Iteration': new_detection, 'Time': start_date, 'Success_flag': 1}, ignore_index=True)
        elif ('[INFO ]' in line) and (flag==0): #Missed Detection
            flag=1
            missed_detection+=1
            new_detection+=1
            df = df.append({'Iteration': new_detection, 'Time': start_date, 'Success_flag': 1}, ignore_index=True)

print('Total Trains : '+str(new_detection))
print('Number of Successful Train Detection : '+str(success_detection))
print('Number of Missed Train Detection : '+str(missed_detection))       
```

```python
df.to_csv('workbench.csv', sep=',')
```

# Building the model

```python
# Splitting the dataset into features and label
# Stepwise regression #2
columns = ['seconds_diff_with_next_train','Trains_per_hour']
labels = df['Success_flag'].values
features = df[list(columns)].values
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.20)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Fitting Decision Tree Classification to the Training set
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
# Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Decision Tree Classification (Training set)')
plt.xlabel('seconds_diff_with_next_train/Trains_per_hour')
plt.ylabel('Success_flag')
plt.legend()
plt.show()
```

```python
# Visualising the Test set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Decision Tree Classification (Test set)')
plt.xlabel('seconds_diff_with_next_train/Trains_per_hour')
plt.ylabel('Success_flag')
plt.legend()
plt.show()
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import accuracy_score
ac= accuracy_score(y_test, y_pred)
print(ac)
```

```python
# Applying k-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
```
